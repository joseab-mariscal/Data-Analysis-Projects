```python
import yfinance as yf
import pandas as pd
import requests
from bs4 import BeautifulSoup
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
def make_graph(stock_data, revenue_data, stock):
    fig = make_subplots(rows=2, cols=1, shared_xaxes=True, subplot_titles=("Histórico del precio de la acción", "Histórico de ingresos"), vertical_spacing = .3)
    fig.add_trace(go.Scatter(x=pd.to_datetime(stock_data.Date, infer_datetime_format=True), y=stock_data.Close.astype("float"), name="Share Price"), row=1, col=1)
    fig.add_trace(go.Scatter(x=pd.to_datetime(revenue_data.Date, infer_datetime_format=True), y=revenue_data.Revenue.astype("float"), name="Revenue"), row=2, col=1)
    fig.update_xaxes(title_text="Fecha", row=1, col=1)
    fig.update_xaxes(title_text="Fecha", row=2, col=1)
    fig.update_yaxes(title_text="Precio ($US)", row=1, col=1)
    fig.update_yaxes(title_text="Ingresos ($US Millions)", row=2, col=1)
    fig.update_layout(showlegend=False,height=900,title=stock,xaxis_rangeslider_visible=True)
    fig.show()
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
headers = {"User-Agent": "Mozilla/5.0"}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')
tablas = soup.find_all("table")
print(f"Se encontraron {len(tablas)} tablas en la página.")
indice_tabla = None
for i, tabla in enumerate(tablas):
    if "Tesla Quarterly Revenue" in str(tabla):
        indice_tabla = i
        break
if indice_tabla is None:
    print("Error: No se encontró la tabla con 'Tesla Quarterly Revenue'")
else:
    print(f"Tabla encontrada en el índice {indice_tabla}")


ingresos_tesla = pd.DataFrame(columns=["Date", "Revenue"])


for fila in tablas[indice_tabla].find_all("tr")[1:]:  
    columnas = fila.find_all("td")
    if len(columnas) >= 2:  
        fecha = columnas[0].text.strip()
        ingresos = columnas[1].text.strip().replace("$", "").replace(",", "")  
        if ingresos != "":
            ingresos_tesla = ingresos_tesla.append({"Date": fecha, "Revenue": ingresos}, ignore_index=True)


print(ingresos_tesla.tail())
```

```python

url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"


headers = {"User-Agent": "Mozilla/5.0"}


response = requests.get(url, headers=headers)


if response.status_code == 200:
    print("Conexión exitosa.")
else:
    print(f"Error al acceder a la página. Código de estado: {response.status_code}")


soup = BeautifulSoup(response.text, "html.parser")


tablas = soup.find_all("table")
print(f"Se encontraron {len(tablas)} tablas en la página.")


indice_tabla = None
for i, tabla in enumerate(tablas):
    if "Tesla Quarterly Revenue" in str(tabla):  
        indice_tabla = i
        break


if indice_tabla is None:
    print("Error: No se encontró la tabla de ingresos de Tesla.")
else:
    print(f"Tabla encontrada en el índice {indice_tabla}")


datos = []


for fila in tablas[indice_tabla].find_all("tr")[1:]:  
    columnas = fila.find_all("td")
    
    
    if len(columnas) >= 2:
        fecha = columnas[0].text.strip()  
        ingresos = columnas[1].text.strip().replace("$", "").replace(",", "")  
        
        
        if ingresos != "":
            datos.append({"Date": fecha, "Revenue": ingresos})  


ingresos_tesla = pd.DataFrame(datos)


print(ingresos_tesla.tail())


ingresos_tesla.to_csv("tesla_revenue.csv", index=False)
print("Datos guardados en 'tesla_revenue.csv'")
```

```python
import matplotlib.pyplot as plt

def make_graph(dataframe, title):
    """Función para graficar los ingresos de una empresa."""
    plt.figure(figsize=(10, 5))  
    plt.plot(dataframe["Date"], dataframe["Revenue"], marker="o", linestyle="-", color="b", label="Revenue")
    
    plt.xlabel("Fecha")
    plt.ylabel("Ingresos en millones (USD)")
    plt.title(title)
    plt.xticks(rotation=45)  
    plt.legend()
    plt.grid()
    plt.show()
```

```python

ingresos_tesla = pd.read_csv("tesla_revenue.csv")
ingresos_gamestop = pd.read_csv("gamestop_revenue.csv")


ingresos_tesla["Revenue"] = pd.to_numeric(ingresos_tesla["Revenue"], errors="coerce")
ingresos_gamestop["Revenue"] = pd.to_numeric(ingresos_gamestop["Revenue"], errors="coerce")


ingresos_tesla = ingresos_tesla.sort_values(by="Date")
ingresos_gamestop = ingresos_gamestop.sort_values(by="Date")
```

```python

make_graph(ingresos_tesla, "Ingresos de Tesla (Quarterly Revenue)")


make_graph(ingresos_gamestop, "Ingresos de GameStop (Quarterly Revenue)")
```
